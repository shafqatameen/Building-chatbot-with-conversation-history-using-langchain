```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
groq_api_key=os.getenv("GROQ_API_KEY")
```

```python
from langchain_groq import ChatGroq
model=ChatGroq(model="openai/gpt-oss-120b",groq_api_key=groq_api_key

)
model
```

```python
from langchain_groq import ChatGroq
model2=ChatGroq(model="whisper-large-v3-turbo",groq_api_key=groq_api_key

)
model2
```

```python
from langchain_core.messages import HumanMessage 
model.invoke([HumanMessage(content="hi my name is shafqat and i am  AI Engineer")])
```

```python
from langchain_core.messages import AIMessage
model.invoke([
    HumanMessage(content="hi my name is shafqat and i am  AI Engineer"),
    AIMessage(content="Hello Shafqat! 👋 Great to meet an AI Engineer. How’s everything going in the world of machine learning and AI? Anything exciting you’re working on or curious about right now?'"),
    HumanMessage(content="what is my name and what am i doing?")
])
```

```python
config={"configuration":{'session_id','chat1'}}
```

```python
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory

store={}
def get_session_history(session_id:str)->BaseChatMessageHistory:
    if session_id not in store:
        store[session_id]=ChatMessageHistory()
    return store[session_id]

with_session_history_model=RunnableWithMessageHistory(model,get_session_history)
```

```python
config={"configurable":{"session_id":"chat"}}
```

```python
result=with_session_history_model.invoke([
    HumanMessage(content="my name is anwar")],
    config=config
)
```

```python
result.content
```

```python
with_session_history_model.invoke(
    [HumanMessage(content="what is my name?")],
    config=config)
```

```python
##change the config-->>change the session id
config2={'configurable':{"session_id":"chat2"}}
with_session_history_model.invoke(
    [HumanMessage(content="what is my name?")],
    config=config2

)
```

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
prompt=ChatPromptTemplate.from_messages(
    ["system","your a helpful ai asistant, answer best of your knowledge and capabilities", MessagesPlaceholder(variable_name="messages")]
   
)
chain=prompt|model
```

```python
chain.invoke({'messages':[HumanMessage(content="hi my name is shafqat")]})
```

```python
with_session_history_model=RunnableWithMessageHistory(chain,get_session_history)
```

```python
config3={'configurable':{'session_id':'chat3'}}
with_session_history_model.invoke([HumanMessage(content="what is my name?")],config=config3)
```
